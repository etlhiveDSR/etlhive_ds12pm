# Read Dataset

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
import pandas as pd
A = pd.read_csv("Desktop/Vaibhav/Datasets/Restaurant_Reviews.tsv",sep="\t")
```

```python
A.head()
```

# Preprocessing

Q = {
    "went":"go",

    "gone":"go"
}

```python
#Q['went']
```

```python
from re import sub
from nltk.stem import PorterStemmer
ps = PorterStemmer()
P = []
for i in A.Review:
    q = sub("[^a-zA-Z0-9 ]","",i)
    q = q.lower()
    q = ps.stem(q)
    P.append(q)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
Xnew = cv.fit_transform(P).toarray()
```

```python
Xnew
```

```python
pd.set_option("display.max_columns",300)
pd.DataFrame(cv.get_feature_names()).T
```

# Train test split

```python
Y = A[["Liked"]]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=31)
```

# Model

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
model = nb.fit(xtrain,ytrain)
```

# Predictions and Loss

```python
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
from sklearn.metrics import accuracy_score
tr_acc = accuracy_score(ytrain,tr_pred)
ts_acc = accuracy_score(ytest,ts_pred)
```

```python
print("Training accuracy",tr_acc)
print("Testing accuracy",ts_acc)
```
